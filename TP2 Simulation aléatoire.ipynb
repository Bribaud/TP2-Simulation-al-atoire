```r
# Fonction auxiliaire
quatrejets = function(){
unsixouplus <- 0
jeu <- sample(1:6,4,replace=T)
nbsix = sum(jeu == 6)
if(nbsix >= 1) unsixouplus <- 1
return(unsixouplus)
}
# Fonction auxiliaire
vingtquatrejets = function(){
undoublesixouplus <- 0
de1 <- sample(1:6,24,replace=T)
de2 <- sample(1:6,24,replace=T)
nbdoublesix <- sum(de1[which(de1 == de2)] == 6)
if(nbdoublesix >= 1) undoublesixouplus <- 1
return(undoublesixouplus)
}
# Estimation de la probabilité d’un événement
meresix = function(nsim = 2000){
aumoinsunsix <- 0
aumoinsundoublesix <- 0
for(j in 1:nsim){
aumoinsunsix <- aumoinsunsix + quatrejets()
aumoinsundoublesix <- aumoinsundoublesix + vingtquatrejets()
}
#************* Affichage des résultats *******************
cat("Fréquence des six =",aumoinsunsix/nsim,"\n")
cat("Fréquence des doubles six =",aumoinsundoublesix/nsim,"\n")
}
meresix()
```

```r
Freq <- function(n) {
  # Générer n tirages aléatoires d'un dé cubique équilibré
  resultats <- sample(1:6, n, replace = TRUE)
  
  # Calculer la fréquence d'obtenir un 5
  frequence_5 <- sum(resultats == 5) / n
  
  return(frequence_5)
}

# Test de la fonction pour n = 10, 100, 1000
n_valeurs <- c(10, 100, 1000)

# Affichage des résultats
for (n in n_valeurs) {
  frequence <- Freq(n)
  message(paste("Fréquence pour n =", n, ":", frequence))
}
```

```r
# Définir le nombre d'étapes
n <- 100

# Simuler les déplacements (1 pour droite, -1 pour gauche) avec probabilité 0.5
deplacements <- sample(c(1, -1), n, replace = TRUE, prob = c(0.5, 0.5))

# Obtenir la position cumulative à chaque étape
positions <- cumsum(c(0, deplacements))

# Afficher les résultats
print(positions)
```

```r
# Fonction pour simuler un lancer de dé
lancer_de <- function() {
  return(sample(1:6, 5, replace = TRUE))
}

# Initialiser le nombre de lancers
nombre_lancers <- 0

# Simuler le processus jusqu'à ce que tous les dés affichent 6
while(TRUE) {
  des <- lancer_de()
  nombre_lancers <- nombre_lancers + 1
  
  # Afficher les résultats après chaque lancer
  cat("Lancer", nombre_lancers, ":", des, "\n")
  
  # Vérifier le nombre de dés qui n'ont pas affiché un 6
  des_non_six <- sum(des != 6)
  
  if (des_non_six == 0) {
    # Tous les dés affichent 6, sortie de la boucle
    break
  }
}

# Afficher le nombre total de lancers
cat("Nombre total de lancers :", nombre_lancers, "\n")
```
